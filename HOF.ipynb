## Fungsi Tingkat Tinggi

Dalam Python, fungsi diperlakukan sebagai warga negara kelas satu, sehingga Anda dapat melakukan operasi berikut pada fungsi:

- Suatu fungsi dapat mengambil satu atau lebih fungsi sebagai parameter
- Suatu fungsi dapat dikembalikan sebagai hasil dari fungsi lain
- Suatu fungsi dapat dimodifikasi
- Suatu fungsi dapat ditetapkan ke variabel

Di bagian ini, kita akan membahas:1. Menangani fungsi sebagai parameter
2. Mengembalikan fungsi sebagai nilai kembalian dari fungsi lain
3. Menggunakan penutupan dan dekorator Python

## Dekorator PythonDekorator adalah pola desain dengan Python yang memungkinkan pengguna menambahkan fungsionalitas baru ke objek yang sudah ada tanpa mengubah strukturnya. Dekorator biasanya dipanggil sebelum definisi fungsi yang ingin Anda hias.

### Membuat Dekorator

Untuk membuat fungsi dekorator, kita memerlukan fungsi luar dengan fungsi pembungkus dalam.

**Contoh:**

```py
# Fungsi normal
def salam():
    kembalikan 'Selamat datang di Python'
def huruf besar_decorator(fungsi):
    def pembungkus():
        fungsi = fungsi()
        make_uppercase = fungsi.upper()
        kembalikan make_uppercase
    bungkus kembali
g = huruf besar_dekorator(salam)
mencetak(g()) # SELAMAT DATANG DI PYTHON

## Mari kita terapkan contoh di atas dengan dekorator

'''Fungsi dekorator ini adalah fungsi tingkat tinggi
yang mengambil fungsi sebagai parameter'''
def huruf besar_decorator(fungsi):
    def pembungkus():
        fungsi = fungsi()
        make_uppercase = fungsi.upper()
        kembalikan make_uppercase
    bungkus kembali
@uppercase_decorator
def salam():
```

### Python - Fungsi Peta

Fungsi map() adalah fungsi bawaan yang menggunakan suatu fungsi dan dapat diubah sebagai parameter.



### Python - Fungsi Filter

Fungsi filter() memanggil fungsi tertentu yang mengembalikan boolean untuk setiap item dari iterable (daftar) yang ditentukan. Ini menyaring item yang memenuhi kriteria penyaringan.
