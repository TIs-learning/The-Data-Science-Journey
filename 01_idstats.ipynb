Klasifikasi data dalam statistika | data classification

8. Dalam statistika kita mengenal 2 tipe data yaitu data kualitatif dan data kuantitatif 
    - data kualitatif, terdiri atas data non numerik : atribut dan level. contoh : nasi goreng, nasi pecel, nasi timbel
    - data kuantitatif, data numerik yang dihasilkan melalui : perhitungan dan pengukuran. contoh : Rp 10.000. 500. 2.500
9. skala pengukuran : 
- Nominal, : Berasosiasi dengan tipe data kualitatif, berfokus pada pengelompokan data berdasarkan nama, label, atau kualitas. tidak dapat dikenakan label matematika. contoh : merk kendaraan bermotot, kota kelahiran, nama siswa,bulan,hari,nim,plat motor. ini bukan ga boleh angka
- Ordinal, : Berasosiasi dengan tipe data kualitatif,, data dapat dikelompokkan, data dapat disusun berdasarkan urutan, peringkat, atau ranking, tidak dapat dikenakan operasi matematika. contoh tingkat kepuasan pelanggan, temperature suhu, peringkat siswa dikelas
- Interval, : Berasosiasi dengan tipe data kuantitatif, data dapat dikelompokan, data dapat disusun berdasarkan urutan,peringkat,atau ranking, dapat dikenakan operasi matematika untuk menghitung selisih nilai, nilai numerik merepresentasikan posisi pada suatu skala ukur tertentu, tidak memiliki nilai nol yang absolut, tidak dapat dikenakan operasi matematika terkait perkalian. contoh : temperature udara 32 derajat celcius, tahun 2020
- Rasio : Berasosiasi dengan tipe data kuantitatif, data dapat dikelompokan, data dapat disusun berdasarkan urutan,peringkat,atau ranking, dapat dikenakan operasi matematika untuk menghitung selisih nilai, memiliki nilai nol yang absolut, dapat dikenakan operasi matematika terkait perkalian. contoh usian anak 7 tahun, berat badan 63 kg, harga nasi goreng Rp 25.000


Desain Eksperimen dalam statistika | Experimental Design | statistika deskriptif 

10. studi statistik : studi observasi dan studi eksperimen
    - studi observasi : seorang peneliti melakukan pengamatan terhadap subjek (cth : pengukuran karakteristik ) tanpa melakukan tindakan apapun yang dapat mempengaruhi hasil pengamatan
    - studi eksperimen : seorang penliti menerapkan suatu treatment tertentu terhadap subjek sebelum melakukan pengamatan untuk memahami dari efek treatment yang diberikan 
11. Design eksperimen yang diperhatikan :kendali, pengacakan, replikasi
    - kendali (control) : bisa dibagi ke 2 hal yaitu treatment group yaitu subjek dalam experimental study yang menerima treatment sebelum diamati. control grup yaitu subjek dalam eksperimental study yang tidak menerima treatment dan juga diamati , subjek seringkali menerima treatment palsu (placebo/fake treatment)
        - contoh implementasi : blinding yaitu subjek eksperimen tidak mengetahui apakah dirinya menerima treatment atau placebo. double blinding yaitu baik peneliti atau subjek tidak mengetahui apakah seorang subjek menerima treatment atau placebo, dibutuhkan pihak ketiga untuk mendistribusikan treatment dan placebo kepada subjek eksperiment
    - randomisation (pengacakan) : penentuan subjek eksperimen ke dalam treatment group dan control group dilakukan secara acak : experimental unit > pemilihan acak >/ treatment group />control group
        - contoh implementasi : randomised blocked design, keseluruhan subjek eksperimen akan dikelompokan berdasarkan kesamaan karakteristik tertentu (cth: rentang usi, jenis, kelamin), subjek di tiap kelompok lalu dipilih secara acak untuk dimasukan ke dalam treatment group dan control group
    - replikasi (replication) : untuk meningkatkan validitas dari hasil eksperimen, dibutuhkan replikasi atau pengulangan eksperimen dengan kondisi serupa atau mirip, replikasi melibatkan subjek eksperimen yang berbeda, dengan demikian replikasi juga akan memperbesar ukuran dari subjek eksperimen (sample size) yang juga dapat menunjang validitas dari hasil eksperimen
        - contoh implementasi : pengujian produk dengan efek dengan melibatkan placebo dan treatment real
11. Tantangan dalam eksperimental study : 
- confouding / lurking variable : faktor ekternal (dan tidak diperhitungkan sebelumnya) yang berpotensi memberikan pengaruh terhadap hasil eksperimen
- placebo effect : subjek eksperimen memberikan reaksi positif walau subjek tersebut menerima placebo (treatment palsu) 
- hawthorne effect : perubahan perilaku dari subjek eksperimen setelah mengetahui bahwa dirinya terlibat dalam eksperimen

Pengumpulan data dalam statistika | Data Collecion

12. 2 pendekatan pengumpulan data : census dan sampling 
    - census : pengumpulan data dilakukan pada tingkat populasi.alhasil akan diperoleh informasi yang sifatnya lengkap, hanya saja ketika ukuran populasinya sangat besar, maka seringkali censius menjadi pilihan yang sulit dan mahal untuk dilakukan 
    - sampling : pengumpulan data dilakukan pada sub bagian dari populasi. alhasil, informasi yang diperoleh sifatnya tidak lengkap.pendekatan semacam ini cukup umum ditemui dalam studi statistik, disini sample baik adalah sample yang dapat merepresentasikan populasinya, dibutuhkan teknik sampling yang tepat untuk mendapatkan sample yang representatif terhadap populasinya
13. sampling error : karena sample merupakan sub bagian dari populasi, maka selisih atau perbedaan nilai antar data sample dan data populasi akan selalu ada, perbedaan atau selish nilai ini dikenal dengan istilah sampling error, bahkan dengan teknik sampling sampling sebaik apapun, sampling error ini tidak dapat dihindari
14. sampling with/without replacemant, sampling without replacement adalah menjamin satu anggota populasi hanya dapat terpilih satu anggota saja sebagai anggota sample kalau with sebaliknya
15. 5 teknik / teknik umum dalam bidang statistika : simple random sampling, stratified sampling, cluster sampling, systematic samping,convenience sampling
    - simple random sampling: kayak namanya aja random, teknik pengumpulan data dimana semua orang berhak menjadi anggota sampling dan dipilih secara acak
    - stratified sampling : nah kalo ini dia tetap acak tapi semisal ada populasi tertentu dia bakal membagi ke dalam jenis karakteristik yang sama contoh jenis kelamin, golongan yang terbentuk ini disebut dengan strata, baru setelah ada beberapa kelompok dipilih secara cak tuh,nah seberapa banyak nya sampel yang diambil itu mengikuti seberapa besar persentasi strata dengan populasi, katanya ada rumus menentukan samplingnya 
    - cluster sampling : nah kalo ini mirip kayak stratified sampling tapi dia membagi nya ke dalam pengelompokan yang sudah ada contohnya : area geografis, kecamatan, kelurahan, dl. kelompok ini disebut cluster sisa nya sama kayak stratified sampling dipilih acak sample nya juga disesuaikan
    - systematic sampling : ini katanya berdasarkan interval cuma ga bisa kalau pada sampling ini ada pola yang sifatnya konsisten dan sistematis contoh survey kepuasan pelanggan setiap 10 pelanggan. maksudnya bermaslaah ini ketika setiap 10 orang semisal besok pagi dia juga yang di survey
    - convenience sampling : ini katany teknik yang jelek cuma mementingkan kemudahan sehingga rentan terhadap bias
    16. bias ini ada sampling bias ( Sampel tidak representatif), measurment bias(Alat ukur atau cara ukur bermasalah), response bias(Responden nggak jujur / ikut gengsi), confirmation bias(Cenderung nyari data yang ngedukung keyakinan sendiri)

51. ini ada namanya Interquartile Range (IQR) : measure of Variation (pengukuran keberagaman/sebaran data) dengan menselisihkan nilai quartile ketiga dan quartile pertama.rumus nya iqr = q3 - q1, nah iqr ini bisa digunakan untuk mendeteksi outlier, entri data(data point) pada suatu dataset bisa dikategorikan sebagai outlier bila - lebih kecil dari q1 - 1.5(iqr) atau lebih besar dari q3 + 1.5(iqr)
